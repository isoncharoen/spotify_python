### Dynamic programming: overall run-through of all factors

```python
#dynamic coding
listwords = ["dating", "violence", "romantic", "communication", "danceability", "world/life", "night/time", "family/gospel", "like/girls", "valence", "energy", "instrumentalness", "acousticness", "feelings", "loudness"]

for keyword in listwords:
    keyword_counts = data.groupby("release_date")[f"{keyword}"].sum()
    plt.bar(keyword_counts.index, keyword_counts.values) #color="magenta"
    plt.title(f"Frequency of '{keyword}' by year")
    plt.xlabel("Year")
    plt.ylabel(f"Count of {keyword}")
    plt.show()
```

```python
pop_data = data[data["genre"] == "pop"]
sadness_counts = pop_data.groupby("release_date")["sadness"].sum()
plt.bar(sadness_counts.index, sadness_counts.values, color="blue")
plt.title("Frequency of sadness within the pop genre")
plt.xlabel("Year")
plt.ylabel("Sadness counts")
plt.show()

pop_data = data[data["genre"] == "hip hop"]
sadness_counts = pop_data.groupby("release_date")["sadness"].sum()
plt.bar(sadness_counts.index, sadness_counts.values, color="blue")
plt.title("Frequency of sadness within the hip hop genre")
plt.xlabel("Year")
plt.ylabel("Sadness counts")
plt.show()

obscene_counts = data.groupby("release_date")["topic"].apply(lambda x: (x == "obscene").sum())
plt.bar(obscene_counts.index, obscene_counts.values, color="orange")
plt.title("Frequency of obscenity")
plt.xlabel("Year")
plt.ylabel("Obscene counts")
plt.show()
```

```python

listwords = ["dating", "violence", "romantic", "communication", "danceability", 
             "world/life", "night/time", "family/gospel", "like/girls", "valence", 
             "energy", "instrumentalness", "acousticness", "feelings", "loudness"]

songs_per_year = data.groupby("release_date").size()

for keyword in listwords:
    keyword_counts = data.groupby("release_date")[keyword].sum()
    
    keyword_avg = keyword_counts / songs_per_year
    
    plt.bar(keyword_avg.index, keyword_avg.values, color="blue")
    plt.title(f"Average frequency of '{keyword}' per year")
    plt.xlabel("Year")
    plt.ylabel(f"Count of {keyword} (per year)")
    plt.show()
```
